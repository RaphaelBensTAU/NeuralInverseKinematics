```python
import trimesh
import glob
import os
```

```python
obj_files = glob.glob('*.obj')
#obj_files = ['hip-roll.stl', 'hip-yaw.stl']
```

```python
m = trimesh.transformations.reflection_matrix((0,0,0), (1,0,0))
for obj in obj_files:
    mesh = trimesh.load(obj)
    mesh.apply_transform(m)
    mesh.export(os.path.splitext(obj)[0] + '_mirror.obj')
```
